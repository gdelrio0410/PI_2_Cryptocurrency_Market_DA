# EDA tether

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
tether_df = pd.read_csv("../Dataframes/tether_df.csv")
tether_df["date"] = pd.to_datetime(tether_df["date"])
```

```python
fig = px.line(tether_df, x="date", y="prices", color="id",
              title="Precio de la moneda")

fig.show()
```

```python
fig_price_returns = px.line(tether_df, x='date', y='price_diff', title='comparacion de precio en el tiempo de acuredo al dia anterior')
fig_price_returns.show()
```

Este comportamiento nos muestra que tanto cambia el precio de la moneda

```python
fig = px.line(tether_df, x="date", y="market_caps", color="id",
              title="Capitalizacion del mercado de la moneda")

fig.show()
```

se puede obserbar que las lineas son muy similares solo que el precio esta en miles y la capitalizacion del mercado esta en miles de millones

```python
fig_price_returns = px.line(tether_df, x='date', y='market_caps_diff', title='comparacion del la capitalizacion de mercado en el tiempo de acuredo al dia anterior')
fig_price_returns.show()
```

```python
fig = px.line(tether_df, x="date", y="total_volumes", color="id",
              title="Volumen de la moneda")

fig.show()
```

```python
fig_price_returns = px.line(tether_df, x='date', y='total_volumes_diff', title='comparacion de volumen en el tiempo de acuredo al dia anterior')
fig_price_returns.show()
```

```python
# Crear un gráfico de barras de las diferencias de precios en diferentes fechas
fig_price_returns = px.scatter(tether_df, x='prices', y='returns', title='Relación entre Precios y Retornos de la moneda')
fig_price_returns.show()
```

```python
volatility = tether_df["returns"].std()
print(f"Volatilidad: {volatility:.2%}")
```
