```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
bitcoin_df = pd.read_csv("../Dataframes/bitcoin_df.csv")
bitcoin_df["date"] = pd.to_datetime(bitcoin_df["date"])
```

```python
fig = px.line(bitcoin_df, x="date", y="prices", color="id",
              title="Precio de la moneda")

fig.show()
```

```python
fig = px.line(bitcoin_df, x="date", y="market_caps", color="id",
              title="Capitalizacion del mercado de la moneda")

fig.show()
```

se puede obserbar que las lineas son muy similares solo que el precio esta en miles y la capitalizacion del mercado esta en miles de millones
