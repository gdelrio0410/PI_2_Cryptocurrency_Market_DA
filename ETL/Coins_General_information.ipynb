# Eleccion de monedas

```python
from pycoingecko import CoinGeckoAPI
from datetime import datetime
import datetime
import pandas as pd
import myFunctions
```

```python
cg = CoinGeckoAPI()
cg.ping()
```

Elejiremos las top 10 monedas de acuredo con el rango de capitalizacion del mercado

```python
market_cap_usd = cg.get_coins_markets(vs_currency="usd")
market_cap_usd = pd.DataFrame.from_dict(market_cap_usd).sort_values('market_cap_rank')
market_cap_usd_top10 = market_cap_usd.head(10)
market_cap_usd_top10
```

```python
coin_names = market_cap_usd_top10["id"].tolist()
coin_data_list = [myFunctions.get_coin_data(coin_name) for coin_name in coin_names]
# Crear el DataFrame final
df_top10_coins = myFunctions.create_combined_dataframe(coin_data_list)
df_top10_coins
```

```python
rows_to_drop = ["platforms","public_notice","additional_notices","description","links","image","country_origin","market_data","community_data","developer_data","public_interest_stats","status_updates","tickers","ico_data", "contract_address"]
df_top10_coins_cleaned = df_top10_coins.drop(index=rows_to_drop)
df_top10_coins_cleaned
```

```python
df_top10_coins_cleaned.to_csv("../Dataframes/df_top10_coins_cleaned.csv", index=False)
```
