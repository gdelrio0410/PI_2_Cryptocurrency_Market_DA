# Informacion de las monedas seleccionadas

```python
from pycoingecko import CoinGeckoAPI
from datetime import datetime
import matplotlib.pyplot as plt
import time
import pandas as pd
import myFunctions
```

```python
cg = CoinGeckoAPI()
```

```python
start_time = myFunctions.unix_time(2000, 1, 1, 0, 0)
end_time = myFunctions.unix_time(2023, 8, 22, 0, 0)
```

# bitcoin

```python
bitcoin_ohlc = myFunctions.coin_ohlc("bitcoin")

bitcoin_coin_market_data = myFunctions.coin_market_chart_data("bitcoin", start_time, end_time)
bitcoin_market_data_df = myFunctions.get_coin_market_dataframe(bitcoin_coin_market_data)
bitcoin_market_data_df['id'] = 'bitcoin'

# Convertir la columna 'prices' a tipo float
bitcoin_market_data_df['prices'] = bitcoin_market_data_df['prices'].astype(float)
bitcoin_market_data_df['market_caps'] = bitcoin_market_data_df['market_caps'].astype(float)
bitcoin_market_data_df['total_volumes'] = bitcoin_market_data_df['total_volumes'].astype(float)

# Calcular la diferencia de prices, market_cap y total_volumes entre el dÃ­a actual y el dÃ­a anterior
bitcoin_market_data_df['price_diff'] = bitcoin_market_data_df['prices'] - bitcoin_market_data_df['prices'].shift(1)
bitcoin_market_data_df['market_caps_diff'] = bitcoin_market_data_df['market_caps'] - bitcoin_market_data_df['market_caps'].shift(1)
bitcoin_market_data_df['total_volumes_diff'] = bitcoin_market_data_df['total_volumes'] - bitcoin_market_data_df['total_volumes'].shift(1)

bitcoin_market_data_df

# Convertir las columnas de fecha a tipo datetime
bitcoin_market_data_df['timestamp'] = pd.to_datetime(bitcoin_market_data_df['timestamp'])
bitcoin_ohlc['date'] = pd.to_datetime(bitcoin_ohlc['date'])

# Unir los DataFrames utilizando merge
bitcoin_df = pd.merge(bitcoin_market_data_df, bitcoin_ohlc, left_on='timestamp', right_on='date', how='inner')

columns_to_drop = ['timestamp', "close"]
bitcoin_df.drop(columns=columns_to_drop, inplace=True)

# Reorganizar columnas
new_order = ['date', 'id', 'prices', 'price_diff', 'market_caps', 'market_caps_diff', 'total_volumes', 'total_volumes_diff', 'date', 'open', 'high', 'low']
bitcoin_df = bitcoin_df[new_order]

# Exportar DataFrame a un archivo CSV
bitcoin_df.to_csv('../Dataframes/bitcoin_df.csv', index=False)

bitcoin_df
```
